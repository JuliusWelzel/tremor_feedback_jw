# Statistical Power Analysis
The sample size for the sequential design was estimated based on Archer ea.

## Power analysis

We can compute a power analysis using functions from the `statsmodels.stats.power` package. Let's determine a difference in the mean between two groups. We can compute the sample size needed:

```python
import numpy as np
import statsmodels.stats.power as smp
import matplotlib.pyplot as plt
import math

pa_tt = smp.TTestPower()
sample_size = pa_tt.solve_power(effect_size=0.728, power=0.8, alpha=0.05)


print(f"Based on Archer ea., the minimal sample size to detect differences for the force data, should be {sample_size :2.0f} ")
```

```python
sample_sizes = np.array(range(5, 30, 2))

plt.style.use('seaborn')
fig = plt.figure(dpi=600)
ax = fig.add_subplot(1, 1, 1)
fig = pa_tt.plot_power(
    dep_var='nobs', nobs=sample_sizes,  
    effect_size=[0.728], alpha=0.05, ax=ax, 
    title='Power of Samples t-test\n$\\alpha = 0.05$')
ax.set_ylabel('Power')
ax.axvline(17,label='Estimated n', color='black')
ax.axvline(14,label='Actual n', color='blue')
ax.legend()
```
